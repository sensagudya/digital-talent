```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

1. Prepare a list of offices sorted by country, state, city

```python
run("SELECT city,state,country FROM offices")
```

2. How many employees are there in the company?

```python
run("SELECT COUNT(*) AS totalEmployees FROM employees")
```

3. What is the total of payments received?

```python
run("SELECT SUM(amount) AS Amount FROM payments")
```

4. List the product lines that contain 'Cars'

```python
run("SELECT productLine,textDescription FROM productlines WHERE productLine LIKE '%Cars%'")
```

5. Report total payments for October 28, 2004

```python
run("SELECT paymentDate,SUM(amount) AS Amount FROM payments WHERE paymentDate = '2004-10-28'")
```

6. Report those payments greater than $100,000

```python
run("SELECT checkNumber,amount FROM payments WHERE amount > 100000")
```

7. List the products in each product line

```python
run("SELECT productLine,productName FROM products ORDER BY productLine")
```

8. How many products in each product line?

```python
run("SELECT productLine,COUNT(productName) as Total FROM products GROUP BY productLine")
```

9. What is the minimum payment received?

```python
run("SELECT checkNumber,MIN(amount) AS minimumPayment FROM payments")
```

10. List all payments greater than twice the average payment

```python
run("SELECT checkNumber,amount FROM payments WHERE amount > 2 * (SELECT AVG(amount) FROM payments)")
```

11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT AVG(FORMAT(buyPrice,'P')) AS percentagePrice FROM products")
```

12. How many distinct products does ClassicModels sell?

```python
run("SELECT DISTINCT COUNT(productName) FROM products ")
```

13. Report the name and city of customers who don't have sales representatives

```python
run("SELECT customerName,city FROM customers WHERE salesRepEmployeeNumber IS NULL")
```

14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting

```python
run("SELECT CONCAT(firstName,' ',lastName) AS employeeName,jobTitle FROM employees WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'")
```

15. Which orders have a value greater than $5,000?

```python
run("SELECT productCode,(quantityOrdered * priceEach) AS Amount FROM orderdetails WHERE (quantityOrdered * priceEach) > 5000")
```
