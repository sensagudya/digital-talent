# Analisis data_1 sebagai APBN Pendapatan

```python
# grouping data 1
data1_grouped = data_1.groupby(['tahun','sumber_pendapatan','keterangan_sumber_pendapatan'])
```

# FINAL ANALYSIS

Berdasarkan analisis kedua jenis data APBN, yaitu APBN Pendapatan dan Pembelanjaan, periode 1990 - 2013, diperoleh beberapa hasil sebagai berikut :

1. Sumber pendapatan berbentuk pajak, bukan pajak, dan hibah, dimana masing-masing memiliki jenisnya masing-masing. Jenis bukan pajak adalah Bagian Laba BUMN, PNBP lainnya, Pendapatan Badan Layanan Umum, SDA (Gas Alam, Minyak Bumi, Sumber Daya Alam lainnya), dan surplus Bank Indonesia. Jenis hibah hanya dikategorikan satu yaitu hibah. Jenis pajak berupa pajak dalam negeri (BPHTB, Cukai, PBB, PPH Migas, PPH Non Migas, PPN, Pajak lainnya) dan pajak perdagangan internasional (bea masuk, pajak ekspor)
2. Jumlah pendapatan sepanjang tahun adalah 10894639,3 milyar rupiah.
3. Pendapatan tertinggi didapatkan di tahun 2013 pada sumber pendapatan pajak dalam negeri (PPH Non Migas) sebanyak 513509 milyar rupiah.
4. Jenis pendapatan PPH Non Migas selalu ada pada setiap tahun, tetapi mengalami penurunan pada rentang tahun 1998 - 1999 yang kemudian meningkat pesat di tahun 2000.
5. Pos belanja berbentuk anggaran untuk daerah, pemerintah pusat, dan suspen dimana masing-masing memiliki jenisnya masing-masing. Jenis anggaran untuk daerah adalah dana otonomi khusus dan penyeimbangan, dana perimbangan. Jenis suspen hanya dikategorikan satu yaitu suspen. Jenis anggaran untuk pemerintah pusat berupa pengeluaran pembangunan dan pengeluaran rutin, yang mana terdiri dari banyak tipe pengeluaran rutin.
6. Jumlah pendapatan sepanjang tahun adalah 8602820,7 milyar rupiah.
7. Pembelanjaan tertinggi dilakukan pada anggaran belanja untuk daerah pada jenis dana perimbangan sebagai dana alokasi umum.
8. Pada tahun 1999, dana alokasi umum ini mengalami penyusutan dibandingkan dengan tahun-tahun sebelumnya yang terus meningkat. Akan tetapi, tahun 2000, dana alokasi umum kembali meningkat hingga 2013.
9. Melihat total pendapatan dan pembelanjaan di tahun 1990 - 2013, selisih yang didapatkan tidak terlalu jauh, hanya mencapai kurang lebih 2 milyar rupiah. 
10. Akan tetapi, selisih yang diperoleh dapat lebih kecil dari itu karena terdapat beberapa data kosong (null) pada APBN belanja sehingga jumlah pembelanjaan diestimasikan dapat meningkat.
